```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv("C:\\Users\\Prashant Singh\\OneDrive\\Documents\\Internships\\Alfido Tech\\Website Traffic_task3\\traffic.csv")

# Data Cleaning
df.drop_duplicates(inplace=True)
df['date'] = pd.to_datetime(df['date'])

# Traffic Trends Analysis
traffic_trend = df.groupby('date').size()
plt.figure(figsize=(12, 6))
plt.plot(traffic_trend.index, traffic_trend.values, marker='o')
plt.title('Traffic Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Page Views')
plt.grid(True)
plt.show()

# Top-Performing Pages
top_pages = df['track'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_pages.values, y=top_pages.index)
plt.title('Top 10 Performing Tracks')
plt.xlabel('Number of Page Views')
plt.ylabel('Track')
plt.show()

# Geographic Distribution
geo_dist = df.groupby(['country', 'city']).size().unstack().fillna(0)
plt.figure(figsize=(12, 6))
sns.heatmap(geo_dist, cmap='coolwarm', linewidths=0.5)
plt.title('Geographic Distribution of Traffic')
plt.show()
```

```python
import pandas as pd

# Load the dataset
file_path = r"C:\\Users\\Prashant Singh\\OneDrive\\Documents\\Internships\\Alfido Tech\\Website Traffic_task3\\traffic.csv"  # Replace with the actual file path
data = pd.read_csv(file_path)

# Preview the data
print("Dataset Preview:")
print(data.head())

# Check the data types and missing values
print("\nData Types:")
print(data.dtypes)
print("\nMissing Values:")
print(data.isnull().sum())
```

```python
# Handle missing values
data.fillna(method='ffill', inplace=True)  # Forward fill for missing values

# Convert 'date' column to datetime
data['date'] = pd.to_datetime(data['date'], errors='coerce')

# Drop rows with any remaining missing values
data.dropna(inplace=True)

# Confirm data cleaning
print("\nData after Cleaning:")
print(data.info())
```

```python
# Group by date to see daily traffic trends
daily_traffic = data.groupby('date').size()

print("\nDaily Traffic:")
print(daily_traffic.head())

# Plot the daily traffic trends
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
daily_traffic.plot()
plt.title('Daily Traffic Trends')
plt.xlabel('Date')
plt.ylabel('Number of Page Views')
plt.grid(True)
plt.show()
```

```python
# Analyze the most popular pages or events
top_pages = data['event'].value_counts().head(10)

print("\nTop 10 Most Popular Pages/Events:")
print(top_pages)

# Plot the top pages
plt.figure(figsize=(10, 6))
top_pages.plot(kind='bar')
plt.title('Top 10 Most Popular Pages/Events')
plt.xlabel('Page/Event')
plt.ylabel('Number of Views')
plt.xticks(rotation=45)
plt.show()
```

```python
# Analyze traffic sources if applicable
if 'source' in data.columns:
    traffic_sources = data['source'].value_counts()

    print("\nTraffic Sources:")
    print(traffic_sources)

    # Plot traffic sources
    plt.figure(figsize=(10, 6))
    traffic_sources.plot(kind='pie', autopct='%1.1f%%', startangle=140)
    plt.title('Traffic Sources Distribution')
    plt.ylabel('')
    plt.show()
```

```python
# Analyze traffic by country
traffic_by_country = data['country'].value_counts().head(10)

print("\nTop 10 Countries by Traffic:")
print(traffic_by_country)

# Plot traffic by country
plt.figure(figsize=(10, 6))
traffic_by_country.plot(kind='bar')
plt.title('Top 10 Countries by Traffic')
plt.xlabel('Country')
plt.ylabel('Number of Views')
plt.xticks(rotation=45)
plt.show()
```
